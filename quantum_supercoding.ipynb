```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, transpile, assemble, execute
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

```python
qubit_1 = QuantumRegister(1, "qubit_1")
qubit_2 = QuantumRegister(1, "qubit_2")

c1 = ClassicalRegister(1, "c1")
c2 = ClassicalRegister(1, "c2")

quantum_circut = QuantumCircuit(qubit_2, qubit_1, c2, c1)
```

```python
## Entangling Qubits

quantum_circut.h(qubit_1)
quantum_circut.cx(qubit_1, qubit_2)
```

```python
## Encode data to QUBITS:

## 00 -> I
## 01 -> Z
## 10 -> X
## 11 -> XZ

# quantum_circut.z(qubit_1)
quantum_circut.x(qubit_1)
```

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(quantum_circut, backend).result()
psi = result.get_statevector()

print("Status of Q-bits after teleporting:")
display(plot_bloch_multivector(psi))
```

```python
quantum_circut.cx(qubit_1, qubit_2)
quantum_circut.h(qubit_1)
quantum_circut.measure(qubit_1, c1)
quantum_circut.measure(qubit_2, c2)
```

```python
display(quantum_circut.draw(output="mpl", reverse_bits=True))
```
